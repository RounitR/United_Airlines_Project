```python
# Importing the pandas library, which is essential for data analysis
import pandas as pd

# Loading the datasets into DataFrames using pandas
calls = pd.read_csv('calls.csv')
customers = pd.read_csv('customers.csv')
reason = pd.read_csv('reason.csv')
sentiment_statistics = pd.read_csv('sentiment_statistics.csv')
test = pd.read_csv('test.csv')

# Display the first few rows of each dataset to understand its structure
print(calls.head())
print(customers.head())
print(reason.head())
print(sentiment_statistics.head())
print(test.head())
```

```python
# Display information about each dataset to understand its structure
print("Calls Dataset:")
calls.info()
print("\nCustomers Dataset:")
customers.info()
print("\nReason Dataset:")
reason.info()
print("\nSentiment Statistics Dataset:")
sentiment_statistics.info()
print("\nTest Dataset:")
test.info()
```

```python
# Check for missing values in each dataset and print the results
print("Missing values in Calls Dataset:")
print(calls.isnull().sum())

print("\nMissing values in Customers Dataset:")
print(customers.isnull().sum())

print("\nMissing values in Reason Dataset:")
print(reason.isnull().sum())

print("\nMissing values in Sentiment Statistics Dataset:")
print(sentiment_statistics.isnull().sum())

print("\nMissing values in Test Dataset:")
print(test.isnull().sum())
```

```python
# Example of filling missing values in the 'calls' dataset with 'Unknown'
calls = calls.fillna('Unknown')

# Display the first few rows to check the result of missing value handling
print(calls.head())
print(customers.head())

```

```python
# Drop duplicate rows from each dataset
calls = calls.drop_duplicates()
customers = customers.drop_duplicates()
reason = reason.drop_duplicates()
sentiment_statistics = sentiment_statistics.drop_duplicates()
test = test.drop_duplicates()

# Display the shape of each dataset to verify the duplicates were removed
print("Calls dataset shape:", calls.shape)
print("Customers dataset shape:", customers.shape)
print("Reason dataset shape:", reason.shape)
print("Sentiment Statistics dataset shape:", sentiment_statistics.shape)
print("Test dataset shape:", test.shape)
```

```python
# Convert the relevant columns to datetime format
calls['call_start_datetime'] = pd.to_datetime(calls['call_start_datetime'])
calls['agent_assigned_datetime'] = pd.to_datetime(calls['agent_assigned_datetime'])
calls['call_end_datetime'] = pd.to_datetime(calls['call_end_datetime'])
```

```python
# Calculate the total handle time (time from when the agent answered to when the call ended)
calls['handle_time'] = (calls['call_end_datetime'] - calls['agent_assigned_datetime']).dt.total_seconds()

# Calculate the total waiting time (time from when the call started to when the agent answered)
calls['waiting_time'] = (calls['agent_assigned_datetime'] - calls['call_start_datetime']).dt.total_seconds()
```

```python
# Calculate Average Handle Time (AHT)
total_handle_time = calls['handle_time'].sum()
total_calls = len(calls)
AHT = total_handle_time / total_calls

# Calculate Average Speed to Answer (AST)
total_waiting_time = calls['waiting_time'].sum()
AST = total_waiting_time / total_calls

# Display the results
print(f'Average Handle Time (AHT): {AHT:.2f} seconds')
print(f'Average Speed to Answer (AST): {AST:.2f} seconds')
```

```python
# Group the data by agent_id to analyze their performance
agent_performance = calls.groupby('agent_id')['handle_time'].mean().sort_values(ascending=False)

# Display the agents with the highest and lowest average handle times
print("Top 5 agents with the highest AHT:")
print(agent_performance.head(5))
print("\nTop 5 agents with the lowest AHT:")
print(agent_performance.tail(5))
```

```python
# Load the reason.csv file
reason_data = pd.read_csv('reason.csv')

# Display the first few rows of the dataset to understand its structure
print(reason_data.head())

# Display the column names to ensure that 'primary_call_reason' is present
print(reason_data.columns)
```
